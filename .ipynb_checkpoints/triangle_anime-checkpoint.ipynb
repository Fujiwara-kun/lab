```python
from vpython import *
import math

scene.title = "三角形辺長制約アニメーション（反転付き）"
scene.background = color.white
scene.center = vector(0,0,0)

v1 = vector(2, 0, 0)

length_v0v1_min = 0
length_v0v1_max = 3.0
length_v0v1 = length_v0v1_min

length_v0v2 = 2.0
length_v2v1 = 2.0

v0 = vector(v1.x - length_v0v1, v1.y, 0)

s_v0 = sphere(pos=v0, radius=0.05, color=color.red)
s_v1 = sphere(pos=v1, radius=0.05, color=color.blue)
s_v2 = sphere(pos=vector(0,0,0), radius=0.05, color=color.green)

line_v0v1 = curve(color=color.black, radius=0.01)
line_v0v2 = curve(color=color.black, radius=0.01)
line_v2v1 = curve(color=color.black, radius=0.01)

def circle_circle_intersection(c1, r1, c2, r2):
    dx = c2.x - c1.x
    dy = c2.y - c1.y
    d = math.sqrt(dx*dx + dy*dy)
    if d > r1 + r2 or d < abs(r1 - r2) or d == 0:
        return None
    a = (r1*r1 - r2*r2 + d*d) / (2*d)
    h = math.sqrt(r1*r1 - a*a)
    x2 = c1.x + a * dx / d
    y2 = c1.y + a * dy / d
    rx = -dy * (h / d)
    ry = dx * (h / d)
    p1 = vector(x2 + rx, y2 + ry, 0)
    p2 = vector(x2 - rx, y2 - ry, 0)
    return p1, p2

intersections = circle_circle_intersection(v0, length_v0v2, v1, length_v2v1)
v2 = intersections[0] if intersections else vector(0,0,0)

# 速度（距離変化量）
velocity = 0.01

while True:
    rate(60)

    length_v0v1 += velocity
    if length_v0v1 > length_v0v1_max or length_v0v1 < length_v0v1_min:
        velocity = -velocity  # 反転

    v0 = vector(v1.x - length_v0v1, v1.y, 0)
    s_v0.pos = v0

    intersections = circle_circle_intersection(v0, length_v0v2, v1, length_v2v1)
    if intersections:
        v2 = intersections[0]
        s_v2.pos = v2

    line_v0v1.clear()
    line_v0v2.clear()
    line_v2v1.clear()

    line_v0v1.append(pos=v0)
    line_v0v1.append(pos=v1)

    line_v0v2.append(pos=v0)
    line_v0v2.append(pos=v2)

    line_v2v1.append(pos=v2)
    line_v2v1.append(pos=v1)
```
